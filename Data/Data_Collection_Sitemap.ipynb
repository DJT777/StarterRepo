```python
import requests
from bs4 import BeautifulSoup

import requests
from bs4 import BeautifulSoup, Comment

def fetch_text_from_webpage(url):
    """Fetch the visible text from a webpage."""
    response = requests.get(url)
    response.raise_for_status()

    soup = BeautifulSoup(response.content, 'html.parser')

    # Remove all script, style, and other unwanted tags
    for script in soup(["script", "style", "head", "title", "meta"]):
        script.extract()

    # Extract visible text, excluding comments
    texts = [element.strip() for element in soup.stripped_strings if type(element) != Comment]

    return " ".join(texts)


def save_text_to_file(text, filename):
    """Save a given text into a .txt file."""

    with open(filename, 'w', encoding='utf-8') as file:
        file.write(text)
```

```python
import os
import zipfile

def save_to_zip(batch, base_dir):
    zip_name = base_dir + f"batch_{batch}.zip"
    with zipfile.ZipFile(zip_name, 'w') as z:
        for root, _, files in os.walk(base_dir):
            for file in files:
                if file.endswith(".txt"):
                    file_path = os.path.join(root, file)
                    z.write(file_path, file)
                    os.remove(file_path)
    # Download the created zip file in Google Colab

def save_text_from_webpage(admissions_urls):
    # Ensure the 'cleanedData' directory exists
    if not os.path.exists("cleanedData"):
        os.makedirs("cleanedData")

    batch_size = 50
    current_batch = 1
    file_count = 0

    for url in admissions_urls:
        try:
            cleanedText = fetch_text_from_webpage(url)

            # Just use the significant part of the URL for the filename
            filename_part = url.split("usi.edu/")[1].replace("/", "_")
            filename = "cleanedData/" + filename_part + ".txt"

            with open(filename, "w") as f:
                cleanedText = cleanText(cleanedText)
                textToSave = "URL: " + str(url) +"\n" + filename + "\n" + "\nURL: " + str(url) +"\n" + cleanedText + "\nURL: " + url
                f.write(textToSave)
                print("Saved to FILENAME: " + filename)
                print(textToSave)

            file_count += 1

            if file_count == batch_size:
                save_to_zip(current_batch, "cleanedData/")
                current_batch += 1
                file_count = 0

        except:
            continue

    # Save remaining files (if there's a batch less than 250 at the end)
    if file_count > 0:
        save_to_zip(current_batch, "cleanedData/")
```
